```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_original = cv2.imread('imagenes/imagen1.jpg') 

# Coordenadas asignadas
y1 = 600  
y2 = 900  
x1 = 600  
x2 = 850 

# recortar la parte que se desea
img_recortada = img_original[y1:y2, x1:x2]

cv2.imwrite('imagen_recortada.jpg', img_recortada)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_recortada, cv2.COLOR_BGR2RGB))
plt.title('Imagen Recortada')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

alto = 200
ancho = 200

degradado = np.zeros((alto, ancho, 3), dtype=np.uint8)

color_inicio = np.array([255, 0, 0])  
color_fin = np.array([0, 0, 255])     


for i in range(alto):
    degradado[i, :] = (color_inicio * (alto - i) + color_fin * i) // alto

# Mostrar el degradado
plt.imshow(degradado)
plt.axis('off')
plt.show()
```

Ejercicio 3:

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones del tablero de ajedrez
alto = 800
ancho = 800

# Tamanio de cada casilla
tamano_casilla = 100

# Crear una matriz vacia
tablero = np.zeros((alto, ancho, 3), dtype=np.uint8)

# Colores de las Casillas
color1 = np.array([255, 255, 255])  
color2 = np.array([0, 0, 0])        

# Rellenado de las casillas
for i in range(0, alto, tamano_casilla):
    for j in range(0, ancho, tamano_casilla):
        if (i // tamano_casilla + j // tamano_casilla) % 2 == 0:
            tablero[i:i+tamano_casilla, j:j+tamano_casilla] = color1
        else:
            tablero[i:i+tamano_casilla, j:j+tamano_casilla] = color2

plt.imshow(tablero)
plt.axis('on')
plt.show()
```
