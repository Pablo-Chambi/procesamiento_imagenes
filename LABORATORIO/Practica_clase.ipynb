```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread('riko.jpg', 0)
um_adap = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 11)
umb_adap_gauss = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 11)
um_adap_binario = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 11)
um_adap_binario_inv = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 11)

plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(um_adap, cmap='gray')
plt.title('Umbral Adaptativo (Media)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(umb_adap_gauss, cmap='gray')
plt.title('Umbral Adaptativo (Gaussiano)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(um_adap_binario, cmap='gray')
plt.title('Umbral Adaptativo Binario')
plt.axis('off')

plt.show()
plt.figure(figsize=(5, 5))
plt.imshow(um_adap_binario_inv, cmap='gray')
plt.title('Umbral Adaptativo Binario Inverso')
plt.axis('off')
plt.show()

def histograma(ima):
    histo = cv2.calcHist([ima],[0],None,[256],[0,256])

    plt.figure(figsize = (10,5))
    plt.plot(histo,color = "black")
    plt.title('Histograma')
    plt.xlabel('Intensidad_Pixel')
    plt.ylabel('frecuencia')

ima = cv2.imread('riko.jpg')

histograma(ima)
```
