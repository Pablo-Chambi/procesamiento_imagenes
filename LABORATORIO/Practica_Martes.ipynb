```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

img = cv2.imread('imagenes/riko.jpg', cv2.IMREAD_GRAYSCALE)

histograma, _ = np.histogram(img, 256, [0, 256])
funcion_dist = histograma.cumsum()

hst_normalizado = funcion_dist * histograma.max() / funcion_dist.max()  # Resultado de la funcin
img_ecualizada = cv2.equalizeHist(img)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')

plt.subplot(2, 2, 3)
plt.plot(histograma, color='b')
plt.title('Histograma Original')

plt.subplot(2, 2, 4)
plt.plot(hst_normalizado, color='r')  
plt.title('Histograma Ecualizado')

plt.tight_layout()
plt.show()
```

Algoritmo usando clahe

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

img = cv2.imread('imagenes/riko.jpg', cv2.IMREAD_GRAYSCALE)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_nueva = clahe.apply(img)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(img_nueva, cmap='gray')
plt.title('Imagen con CLAHE')

plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

# Leer la imagen en escala de grises
img = cv2.imread('imagenes/paisaje_casa.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_clahe = clahe.apply(img)

# Ecualizar histograma
img_ecualizada = cv2.equalizeHist(img)

# Calcular los histogramas
hist_original, bins_original = np.histogram(img.flatten(), 256, [0,256])
hist_clahe, bins_clahe = np.histogram(img_clahe.flatten(), 256, [0,256])
hist_ecualizada, bins_ecualizada = np.histogram(img_ecualizada.flatten(), 256, [0,256])

# Mostrar las imgenes y los histogramas
plt.figure(figsize=(15, 10))

# Mostrar la imagen original y su histograma
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(2, 3, 4)
plt.plot(hist_original, color='b')
plt.title('Histograma Original')

plt.subplot(2, 3, 2)
plt.imshow(img_clahe, cmap='gray')
plt.title('Imagen con CLAHE')

plt.subplot(2, 3, 5)
plt.plot(hist_clahe, color='r')
plt.title('Histograma CLAHE')

# Mostrar la imagen ecualizada y su histograma
plt.subplot(2, 3, 3)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')

plt.subplot(2, 3, 6)
plt.plot(hist_ecualizada, color='g')
plt.title('Histograma Ecualizado')

plt.tight_layout()
plt.show()
```

Ecualizacion a Colores

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

# Leer la imagen a color
img = cv2.imread('imagenes/paisaje_casa.jpg', cv2.IMREAD_COLOR)

# Convertir la imagen a escala de grises
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_clahe = clahe.apply(img_grayscale)

# Calcular los histogramas
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_grayscale = cv2.calcHist([img_grayscale], [0], None, [256], [0, 256])
hist_clahe = cv2.calcHist([img_clahe], [0], None, [256], [0, 256])

# Mostrar las imgenes y los histogramas
plt.figure(figsize=(20, 12))

# Imagen Original
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(2, 3, 4)
plt.plot(hist_original, color='b')
plt.title('Histograma de la Imagen Original')
plt.xlim([0, 256])

# Imagen en Escala de Grises
plt.subplot(2, 3, 2)
plt.imshow(img_grayscale, cmap='gray')
plt.title('Imagen en Escala de Grises')

plt.subplot(2, 3, 5)
plt.plot(hist_grayscale, color='g')
plt.title('Histograma de la Imagen en Escala de Grises')
plt.xlim([0, 256])

# Imagen con CLAHE
plt.subplot(2, 3, 3)
plt.imshow(img_clahe, cmap='gray')
plt.title('Imagen con CLAHE')

plt.subplot(2, 3, 6)
plt.plot(hist_clahe, color='r')
plt.title('Histograma de la Imagen con CLAHE')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
