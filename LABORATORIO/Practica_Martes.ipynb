```python
import cv2
import matplotlib.pyplot as plt

# Leer la imagen original
original_img = cv2.imread('riko.jpg')
original_img_rgb = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

# Definir las paletas de colores disponibles en OpenCV junto con sus nombres
paletas = [(cv2.COLORMAP_AUTUMN, 'Autumn'), (cv2.COLORMAP_BONE, 'Bone'), (cv2.COLORMAP_COOL, 'Cool'),
           (cv2.COLORMAP_HOT, 'Hot'), (cv2.COLORMAP_HSV, 'HSV'), (cv2.COLORMAP_JET, 'Jet'),
           (cv2.COLORMAP_OCEAN, 'Ocean'), (cv2.COLORMAP_PARULA, 'Parula'), (cv2.COLORMAP_PINK, 'Pink'),
           (cv2.COLORMAP_RAINBOW, 'Rainbow'), (cv2.COLORMAP_SPRING, 'Spring'), (cv2.COLORMAP_SUMMER, 'Summer'),
           (cv2.COLORMAP_WINTER, 'Winter'), (cv2.COLORMAP_TWILIGHT, 'Twilight'),
           (cv2.COLORMAP_TWILIGHT_SHIFTED, 'Twilight Shifted'), (cv2.COLORMAP_TURBO, 'Turbo'),
           (cv2.COLORMAP_VIRIDIS, 'Viridis'), (cv2.COLORMAP_INFERNO, 'Inferno'),
           (cv2.COLORMAP_MAGMA, 'Magma'), (cv2.COLORMAP_PLASMA, 'Plasma'), (cv2.COLORMAP_CIVIDIS, 'Cividis')]


plt.figure(figsize=(15, 10)) # Configura la ventana de visualización

plt.subplot(4, 6, 1)
plt.imshow(original_img_rgb)
plt.title('Imagen Original')
plt.axis('off')

for i, (paleta, nombre_paleta) in enumerate(paletas):
    img = cv2.applyColorMap(cv2.resize(original_img, (500, 500)), paleta)
    
    plt.subplot(4, 6, i+2)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(nombre_paleta)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
