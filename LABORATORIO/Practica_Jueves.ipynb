Contraste 

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

def contraste(imagen, a=1.0, b=0):
    img_nueva = np.clip(a * imagen + b, 0, 255).astype(np.uint8)
    return img_nueva

imagen_original = cv2.imread('imagenes/riko.jpg')

imagen_gris = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2GRAY)

imagen_con_contraste = contraste(imagen_gris, a=1.5, b=50) 

# Mostrar las imgenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen_gris, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Imagen en escala de grises')

plt.subplot(1, 2, 2)
plt.imshow(imagen_con_contraste, cmap='gray')
plt.title('Imagen con contraste')
plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

def contraste_color(imagen, a=1.0, b=0):
    img_nueva = cv2.convertScaleAbs(imagen, alpha=a, beta=b)
    return img_nueva

imagen = cv2.imread('imagenes/riko.jpg')
blue, red, green = cv2.split(imagen)

contraste_blue = contraste_color(blue, a=1.2, b=1)
contraste_red = contraste_color(red, a=1.2, b=1)
contraste_green = contraste_color(green, a=1.2, b=1)

fin_img = cv2.merge((contraste_blue, contraste_red, contraste_green))

plt.figure(figsize=(15, 4))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(contraste_blue, cv2.COLOR_BGR2RGB))
plt.title('Canal Azul con Contraste')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(contraste_red, cv2.COLOR_BGR2RGB))
plt.title('Canal Rojo con Contraste')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(contraste_green, cv2.COLOR_BGR2RGB))
plt.title('Canal Verde con Contraste')

plt.show()
```

Practica Luminocidad

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
img = Image.open('imagenes/riko.jpg')

# Convertir la imagen a escala de grises
img_gray = img.convert('L')

# Calcula la luminosidad promedio
luminosity = np.mean(img_gray)

# Factor de ajuste para aumentar o disminuir la luminosidad
factor_ajuste = 1.5  # Por ejemplo, aumentar en un 50%

# Ajustar la luminosidad
img_gray_ajustada = img_gray.point(lambda p: min(int(p * factor_ajuste), 255))

# Mostrar la imagen original y la imagen con luminosidad ajustada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(img_gray_ajustada, cmap='gray')
plt.title('Imagen con Luminosidad Ajustada')

plt.show()
```
