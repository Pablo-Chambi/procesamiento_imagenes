```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solarizar(image, umbral):
    img_solarizar = np.where(image < umbral, image, 255 - image)
    return img_solarizar 

def posterizar(image, nivel):
    f_nvl = 255 / nivel
    img_poster = np.round(image / f_nvl) * f_nvl
    return img_poster.astype(np.uint8)

imagen = cv2.imread('imagenes/riko.jpg')
img_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)  # Correccin aqu

umbral = 127
img_sol = solarizar(img_gris, umbral)

nivel = 5
img_posterizada = posterizar(img_gris, nivel)

plt.figure(figsize=(15, 10))

plt.subplot(341)
plt.title('Imagen inicial')
plt.imshow(imagen, cmap="gray")

plt.subplot(342)
plt.title('solarizada')
plt.imshow(img_sol, cmap='gray')

plt.subplot(343)
plt.title('posterizada')
plt.imshow(img_posterizada, cmap='gray')

plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,5]
tamanio_img = 100
R = np.ones((tamanio_img,tamanio_img)) * 2
G = np.ones((tamanio_img,tamanio_img)) * 100
B = np.ones((tamanio_img,tamanio_img)) * 145

imagen = np.dstack((R,G,B))
imagen = np.array(imagen,dtype = int)

plt.figure()
plt.imshow(imagen)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('imagenes/riko.jpg')
negativo = 255 - img

plt.figure()
plt.title('Imagen original en BGR')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure()
plt.title('Negativo en BGR')
plt.imshow(cv2.cvtColor(negativo, cv2.COLOR_BGR2RGB))
plt.show()
```
