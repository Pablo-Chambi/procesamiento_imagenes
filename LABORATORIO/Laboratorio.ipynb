```python
import cv2
import matplotlib.pyplot as plt
```

```python
def mostrar_imagen(imagen, titulo="pingui.png"):

    imagenRGB = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenRGB)
    plt.title(titulo)
    plt.axis('off')
    plt.show()
def procesar_imagen(nombre_archivo):
    imagenBGR = cv2.imread(nombre_archivo)
    mostrar_imagen(imagenBGR, "Imagen en BGR")
    imagenGRIS = cv2.cvtColor(imagenBGR, cv2.COLOR_BGR2GRAY)
    cv2.imwrite("imagen_escala_gris.jpg", imagenGRIS)
    mostrar_imagen(imagenGRIS, "Imagen en Escala de Grises")
nombre_archivo = ('C:/Users/PABLO/Pictures/pinguinos_png-jpeg.jpg')
procesar_imagen(nombre_archivo)
```

```python
import cv2 
import matplotlib.pyplot as plt 

def histograma(imagen):
    histo = cv2.calcHist([imagen],[0],None,[256],[0,256])

    plt.figure(figsize = (10,5))
    plt.plot(histo,color = "black")
    plt.title('Histograma')
    plt.xlabel('Intensidad_Pixel')
    plt.ylabel('frecuencia')
    plt.show()
    
ima = cv2.imread('riko.png',0)
histograma(ima)
    
```

```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread('riko.jpg',0)
_,umbralbi = cv2.threshold(imagen,127,255,cv2.THRESH_BINARY) #convercion de imagen a binario 
_,umbral_inc_bi = cv2.threshold(imagen,127,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize = (10,5))


plt.subplot(1,3,2)
plt.imshow(umbralbi,cmap = 'gray' )
plt.title('binario')
plt.subplot(1,3,3)
plt.imshow(umbral_inc_bi,cmap = 'gray')
plt.title('binario inversa')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Leer la imagen en escala de grises
imagen = cv2.imread('riko.jpg', 0)

# Aplicar umbral adaptativo usando el método de media
um_adap = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 11)

# Aplicar umbral adaptativo usando el método de Gaussiano
umb_adap_gauss = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 11)

# Mostrar las imágenes resultantes
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 2)
plt.imshow(um_adap, cmap='gray')
plt.title('Umbral Adaptativo (Media)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(umb_adap_gauss, cmap='gray')
plt.title('Umbral Adaptativo (Gaussiano)')
plt.axis('off')

plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('riko.jpg', 0)  # para cargar una imagen
imagen2 = np.random.randint(0, 256, (300, 300), dtype=np.uint8)  # escala de grises y llama a una imagen ramdon de 8 bits

_, umb_trunc = cv2.threshold(imagen2, 127, 200, cv2.THRESH_TRUNC)
_, umb_tozero = cv2.threshold(imagen2, 127, 255, cv2.THRESH_TOZERO)
_, umb_tozero_INV = cv2.threshold(imagen2, 127, 255, cv2.THRESH_TOZERO_INV)

plt.figure(figsize=(18, 5))

plt.subplot(141)
plt.imshow(imagen2, cmap='gray')
plt.title('Original')

plt.subplot(142)
plt.imshow(umb_trunc, cmap='gray')
plt.title('Umbrales Trunc')

plt.subplot(143)
plt.imshow(umb_tozero, cmap='gray')
plt.title('Umbrales ToZero')

plt.subplot(144)
plt.imshow(umb_tozero_INV, cmap='gray')
plt.title('Umbrales ToZero Inv')

plt.show()
```
