```python
import cv2
import matplotlib.pyplot as plt
```

```python
def mostrar_imagen(imagen, titulo="pingui.png"):

    imagenRGB = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenRGB)
    plt.title(titulo)
    plt.axis('off')
    plt.show()
def procesar_imagen(nombre_archivo):
    imagenBGR = cv2.imread(nombre_archivo)
    mostrar_imagen(imagenBGR, "Imagen en BGR")
    imagenGRIS = cv2.cvtColor(imagenBGR, cv2.COLOR_BGR2GRAY)
    cv2.imwrite("imagen_escala_gris.jpg", imagenGRIS)
    mostrar_imagen(imagenGRIS, "Imagen en Escala de Grises")
nombre_archivo = ('C:/Users/PABLO/Pictures/pinguinos_png-jpeg.jpg')
procesar_imagen(nombre_archivo)
```
