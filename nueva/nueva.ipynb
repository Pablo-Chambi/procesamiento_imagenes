```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

image = cv2.imread('riko.jpg', cv2.IMREAD_GRAYSCALE)

original_img = image.dtype
print("Profundidad original de color: ", original_img)

# Conversión de la imagen a 16 bits
image_16bits = np.uint16(image)
nueva_img = image_16bits.dtype
print("Nueva profundidad de color: ", nueva_img)

# para guardar la imagen 
cv2.imwrite('image16bits.jpg', image_16bits)
img_guardada = cv2.imread('image16bits.jpg', cv2.IMREAD_UNCHANGED)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('original')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)  # Corregido: Cambié el segundo 1 por 2
plt.title('imagen tratada 16 bits')
plt.imshow(img_guardada, cmap='gray')

plt.show()
```
